```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GroupKFold, RandomizedSearchCV, cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as CVFlow

# --- Load and prepare data ---
train_df = pd.read_csv("Extracted_Columns_Activity_Sleep.csv")
test_df = pd.read_csv("Filtered_Columns_Activity_Sleep_6AM_to_10PM.csv")
if 'ID' in test_df.columns:
    test_df = test_df.rename(columns={'ID': 'Participant'})

for df_ in [train_df, test_df]:
    df_['Time'] = pd.to_datetime(df_['Time'], errors='coerce')
    df_['Hour'] = df_['Time'].dt.hour
    df_['HR_Normalized_to_Speed'] = df_['HR_Normalized'] / (df_['Speed'] + 1e-6)
    df_['HR_Normalized_to_Cadence'] = df_['HR_Normalized'] / (df_['Cadence'] + 1e-6)
    df_['Cadence_to_Speed'] = df_['Cadence'] / (df_['Speed'] + 1e-6)

for win in [5, 7]:
    for col in ['Cadence']:
        train_df[f'{col}_rollmean{win}'] = (
            train_df.groupby('Participant')[col]
            .rolling(window=win, min_periods=1)
            .mean().reset_index(level=0, drop=True)
        )
        test_df[f'{col}_rollmean{win}'] = (
            test_df.groupby('Participant')[col]
            .rolling(window=win, min_periods=1)
            .mean().reset_index(level=0, drop=True)
        )

features = ['HR_Normalized', 'Speed', 'Cadence']
label_col = 'Modified_Activity' if 'Modified_Activity' in train_df.columns else 'Activity4'
train_df = train_df.dropna(subset=features + [label_col, 'Participant'])
test_df = test_df.dropna(subset=features)

le = LabelEncoder()
y = le.fit_transform(train_df[label_col])
X = train_df[features]
groups = train_df['Participant']

# --- SMOTE INSIDE CROSS-VALIDATION using CVFlow ---
cvflow = CVFlow([
    ('smote', SMOTE(random_state=42)),
    ('forest', RandomForestClassifier(random_state=42))
])
param_dist_cv = {
    'forest__n_estimators': [100, 500, 1000, 1500],
    'forest__max_depth': [None, 5, 10, 15, 20],
    'forest__min_samples_leaf': [1, 2, 3, 5, 10],
    'forest__max_features': ['sqrt', 'log2', 0.5, None],
    'forest__bootstrap': [True, False],
    'forest__class_weight': ['balanced', None]
}
gkf = GroupKFold(n_splits=3)
random_search_cv = RandomizedSearchCV(
    cvflow, param_distributions=param_dist_cv,
    n_iter=20, cv=gkf.split(X, y, groups),
    scoring='accuracy', n_jobs=-1, verbose=2, random_state=42
)
random_search_cv.fit(X, y)

print("\nSMOTE Inside CV (No Leakage) - Best params:")
print(random_search_cv.best_params_)
print("Best cross-validated accuracy:", random_search_cv.best_score_)

# --- Cross-val predictions for metrics ---
y_pred_cv = cross_val_predict(random_search_cv.best_estimator_, X, y, cv=gkf, groups=groups)
print(classification_report(y, y_pred_cv, target_names=le.classes_))

# --- Confusion matrix ---
cm = confusion_matrix(y, y_pred_cv)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix (SMOTE Inside CV)')
plt.tight_layout()
plt.show()

# --- ROC curve (basic) ---
lb = LabelBinarizer()
y_bin = lb.fit_transform(y)
n_classes = y_bin.shape[1] if y_bin.ndim > 1 else 1
y_proba_cv = cross_val_predict(random_search_cv.best_estimator_, X, y, cv=gkf, groups=groups, method='predict_proba')
plt.figure(figsize=(8, 6))
colors = plt.cm.get_cmap('tab10', n_classes)
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba_cv[:, i])
    auc_score = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{le.inverse_transform([i])[0]} (AUC={auc_score:.3f})", color=colors(i))
plt.plot([0, 1], [0, 1], "k--", lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC (SMOTE Inside CV)")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# --- ROC Curves with Confidence Intervals (Inside CV) ---
n_folds = 3  # GroupKFold n_splits
mean_fpr = np.linspace(0, 1, 100)
tprs = [[] for _ in range(n_classes)]
aucs = [[] for _ in range(n_classes)]

for train_idx, test_idx in gkf.split(X, y, groups=groups):
    model = random_search_cv.best_estimator_
    model.fit(X.iloc[train_idx], y[train_idx])
    y_test_bin = lb.transform(y[test_idx])
    y_proba = model.predict_proba(X.iloc[test_idx])
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
        interp_tpr = np.interp(mean_fpr, fpr, tpr)
        interp_tpr[0] = 0.0
        tprs[i].append(interp_tpr)
        aucs[i].append(auc(fpr, tpr))

plt.figure(figsize=(10, 8))
cmap = plt.cm.get_cmap('tab10', n_classes)
for i in range(n_classes):
    mean_tpr = np.mean(tprs[i], axis=0)
    std_tpr = np.std(tprs[i], axis=0)
    mean_auc = np.mean(aucs[i])
    std_auc = np.std(aucs[i])
    plt.plot(mean_fpr, mean_tpr, color=cmap(i),
             label=f"{le.inverse_transform([i])[0]} (AUC={mean_auc:.3f} ± {std_auc:.3f})", lw=2)
    plt.fill_between(mean_fpr, np.clip(mean_tpr - std_tpr, 0, 1), np.clip(mean_tpr + std_tpr, 0, 1),
                     color=cmap(i), alpha=0.2)
plt.plot([0, 1], [0, 1], "k--", lw=2)
plt.xlabel("False Positive Rate", fontsize=12, fontweight='bold')
plt.ylabel("True Positive Rate", fontsize=12, fontweight='bold')
plt.title("Multiclass ROC Curves w/ Confidence Intervals (Inside CV)", fontsize=15)
plt.legend(loc="lower right", fontsize=10)
plt.tight_layout()
plt.show()

# --- Feature Importance (smote_cvflow) ---
forest_model = random_search_cv.best_estimator_.named_steps['forest']
importances_cv = forest_model.feature_importances_
indices_cv = np.argsort(importances_cv)
plt.figure(figsize=(8, 6))
plt.barh(np.array(features)[indices_cv], importances_cv[indices_cv])
plt.xlabel("Mean Feature Importance")
plt.title("Random Forest Feature Importances (SMOTE Inside CV)")
plt.tight_layout()
plt.show()

# --- Retrain pipeline on ALL original training data (no SMOTE on test set!) ---
final_cvflow = random_search_cv.best_estimator_
final_cvflow.fit(X, y)
# --- Predict test set ---
y_pred_test_cv = final_cvflow.predict(test_df[features])
test_df['Predicted_Activity'] = le.inverse_transform(y_pred_test_cv)
test_df.to_csv("Predicted_Activity_ForPredic_CVflow.csv", index=False)
print('Saved test set predictions with Predicted_Activity (CVflow) in Predicted_Activity_ForPredic_CVflow.csv')
```
